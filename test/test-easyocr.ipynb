```python
from easyocr import Reader
```

```python
reader = Reader(['vi'])
image_path = r"C:\Users\admin\Downloads\hsg 9 hoai duc.jpg"
result = reader.readtext(image_path, output_format='dict')
```

```python
out = [box['text'] for box in result]

# for (bbox, text, prob) in result:
    # print(f"Detected text: {text} (Confidence: {prob:.2f})")
print(out)
```

```python
import cv2

import matplotlib.pyplot as plt

# Load the image
image = cv2.imread(image_path)

# Draw bounding boxes
for box in result:
    top_left = tuple(map(int, box['boxes'][0]))
    bottom_right = tuple(map(int, box['boxes'][2]))
    image = cv2.rectangle(image, top_left, bottom_right, (0, 255, 0), 1)

# Convert BGR image to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image
plt.figure(figsize=(10, 10))
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```
