```python
import pytesseract
from PIL import Image

# Make sure to specify the path to the tesseract executable if it's not in your PATH
# pytesseract.pytesseract.tesseract_cmd = r'/path/to/tesseract'

# Load an image from file
image = Image.open(r"C:\Users\admin\Downloads\hsg 9 hoai duc.jpg")

# Use pytesseract to do OCR on the image
text = pytesseract.image_to_string(image, lang='vie')

# Print the extracted text
print(text)
```

```python
# Get detailed OCR results, including confidence scores
detailed_data = pytesseract.image_to_data(image, lang='vie', output_type=pytesseract.Output.DICT)

# Filter text with confidence > 0.2
filtered_text = ' '.join([detailed_data['text'][i] for i in range(len(detailed_data['text'])) if float(detailed_data['conf'][i]) > 20])

# Print the filtered text
print(filtered_text)
```

```python
from PIL import Image
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Convert PIL image to OpenCV format
image_cv = np.array(image)
image_cv = cv2.cvtColor(image_cv, cv2.COLOR_RGB2BGR)

# Draw boxes using detailed_data
n_boxes = len(detailed_data['text'])
for i in range(n_boxes):
    if int(detailed_data['conf'][i]) > 20:
        (x, y, w, h) = (detailed_data['left'][i], detailed_data['top'][i], detailed_data['width'][i], detailed_data['height'][i])
        image_cv = cv2.rectangle(image_cv, (x, y), (x + w, y + h), (0, 255, 0), 1)

# Convert back to PIL image to display in Jupyter

image_rgb = cv2.cvtColor(image_cv, cv2.COLOR_RGB2BGR)

plt.figure(figsize=(10, 10))
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```
