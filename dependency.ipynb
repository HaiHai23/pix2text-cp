```python
import os
import modulefinder
from graphviz import Digraph

def find_dependencies(directory):
    finder = modulefinder.ModuleFinder()
    dot = Digraph(comment="Module Dependencies")
    
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith(".py"):
                path = os.path.join(root, file)
                if os.path.getsize(path) > 0:
                    finder.run_script(path)
    
    for module in finder.modules.values():
        for dep in module.globalnames.keys():
            if dep in finder.modules:
                dot.edge(module.__name__, dep)
    
    return dot

# Set your project directory
project_directory = r"D:\FPT\math-ocr\Pix2Text\pix2text"

# Generate and render the graph
dependency_graph = find_dependencies(project_directory)
dependency_graph.render("dependencies", format="png", cleanup=False)
```
